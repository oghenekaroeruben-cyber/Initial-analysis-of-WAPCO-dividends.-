```python
import pandas as pd

scenarios = [
    {'price': 185.00, 'shares': 1000},
    {'price': 190.00, 'shares': 1000},
    {'price': 195.00, 'shares': 1000}
]

df = pd.DataFrame(scenarios)

dividend = 6.00
df['Total_Cost'] = df['price'] * df['shares'] * 1.0135
df['Net_Dividend'] = df['shares'] * dividend * 0.90
df['ROI_Percent'] = (df['Net_Dividend'] / df['Total_Cost']) * 100

print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

scenarios = [
    {'price': 185.00, 'shares': 1000},
    {'price': 190.00, 'shares': 1000},
    {'price': 195.00, 'shares': 1000}
]

df = pd.DataFrame(scenarios)

dividend = 6.00
df['Total_Cost'] = df['price'] * df['shares'] * 1.0135
df['Net_Dividend'] = df['shares'] * dividend * 0.90
df['ROI_Percent'] = (df['Net_Dividend'] / df['Total_Cost']) * 100

# Create a bar chart of ROI
df.plot(kind='bar', x='price', y='ROI_Percent', color='skyblue')
plt.title('WAPCO Dividend ROI by Buy Price')
plt.ylabel('ROI (%)')
plt.show()
```

```python
import pandas as pd

# Correcting the filename to match the uploaded file
df = pd.read_csv('historical_data-Lafarge_Africa.csv')
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y', errors='coerce')
```

```python
# Remove 'M', replace with nothing, and convert to float
df['Vol.'] = df['Vol.'].str.replace('M', '').astype(float) * 1_000_000
print("Volume cleaned!")
```

```python
jan_price = float(df[df['Date'] >= '2026-01-01'].iloc[-1]['Price'])
current_price = float(df.iloc[0]['Price'])
growth = ((current_price - jan_price) / jan_price) * 100
print(f"The stock has grown {growth:.2f}% since the start of the year!")
```

```python
# Ensure 'Price' column is numeric
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Calculate the total Naira value of shares traded each day
# (Price * Volume)
df['Trade_Value_NGN'] = df['Price'] * df['Vol.']

# Show the day with the highest activity
busiest_day = df.sort_values(by='Trade_Value_NGN', ascending=False).iloc[0]

print(f"On {busiest_day['Date'].strftime('%Y-%m-%d')}, people traded ₦{busiest_day['Trade_Value_NGN']:,.2f} worth of WAPCO!")
```
